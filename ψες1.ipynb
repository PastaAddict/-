```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
p=np.pi
```

__Ασκηση 1__

```python
fig, axs = plt.subplots(3, 1, constrained_layout=True ,figsize=(10,10))
fig.suptitle('Ανακατασκευή με χρήση της sinc', fontsize=16)

for i in enumerate([0.1,0.01,0.001]):
    
    Ts=i[1]
    axs[i[0]].set_title('Ts='+str(Ts)+'s')
    axs[i[0]].grid(True)
    '''Διαμορφωση διανυσματος χρονου για καλυτερη απεικονιση'''
    samples=int(1+5/Ts)
    n=np.linspace(0,5,samples)
    
    y_n= np.cos(2*p*n)
    t=np.linspace(0,5,10000)
    axs[i[0]].vlines(n, 0, y_n, color='g') #επιλογη της vlines για αναπαρασταση των ψηφιακων σηματων ως τρενο κρουστικων
    reconstruction=np.sinc(t/Ts)*y_n[0] #αρχικοποιηση του ανακατασκευασμενου σηματος
    
    for n in range(1,samples):
        reconstruction+=np.sinc(t/Ts-n)*y_n[n] #υπολογισμος αθροισματος ανακατασκευης
    axs[i[0]].plot(t,reconstruction)
    
```

```python
'''Ορισμος τετραγωνικου σηματος για την ανακατασκευη'''
def square(t):
    if t>=-0.5 and t<=0.5:
        return 1
    else:
        return 0
'''Οπως παραπανω'''
fig, axs = plt.subplots(3, 1, constrained_layout=True ,figsize=(10,10))
fig.suptitle('Κλιμακωτή ανακατασκευή', fontsize=16)

for i in enumerate([0.1,0.01,0.001]):
    
    Ts=i[1]
    samples=int(1+5/Ts)
    n=np.linspace(0,5,samples)
    axs[i[0]].set_title('Ts='+str(Ts)+'s')
    axs[i[0]].grid(True)
    
    y_n= np.cos(2*p*n)
    t=np.linspace(0,5,10000)
    axs[i[0]].vlines(n, 0, y_n, color='g')
    reconstruction=np.fromiter(map(square,t/Ts),float)*y_n[0]
    
    for n in range(1,samples):
        reconstruction+=np.fromiter(map(square,t/Ts-n),float)*y_n[n]
    axs[i[0]].plot(t,reconstruction)
    
```

```python
'''Οπως παραπανω'''
def trig(t):
    if t>=-1 and t<=1:
        return 1-np.abs(t)
    else:
        return 0
    
fig, axs = plt.subplots(3, 1, constrained_layout=True ,figsize=(10,10))
fig.suptitle('Τριγωνική ανακατασκευή', fontsize=16)

for i in enumerate([0.1,0.01,0.001]):
    
    Ts=i[1]
    samples=int(1+5/Ts)
    n=np.linspace(0,5,samples)
    axs[i[0]].set_title('Ts='+str(Ts)+'s')
    axs[i[0]].grid(True)
    
    y_n= np.cos(2*p*n)
    t=np.linspace(0,5,10000)
    axs[i[0]].vlines(n, 0, y_n, color='g')
    reconstruction=np.fromiter(map(trig,t/Ts),float)*y_n[0]
    
    for n in range(1,samples):
        reconstruction+=np.fromiter(map(trig,t/Ts-n),float)*y_n[n]
    axs[i[0]].plot(t,reconstruction)
    
```
